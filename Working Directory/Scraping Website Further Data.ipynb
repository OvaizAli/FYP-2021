```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
# from selenium.webdriver.support.select import Select
from urllib.request import urlopen
# s=Service(ChromeDriverManager().install())
# driver = webdriver.Chrome(service=s)
# driver.maximize_window()
# driver.get('https://www.google.com')

s=Service(ChromeDriverManager().install())
d = webdriver.Chrome(service=s)
d.get('https://www.timeanddate.com/weather/usa/dayton/historic?month=5&year=2020')
_d = {}
for i in d.find_element_by_id('wt-his-select').find_elements_by_tag_name('option'):
#     print(d.page_source)
    i.click()
    with get_weather_data(d.page_source, False) as weather:
        _d[i.text] = weather
    

d.close()
```
